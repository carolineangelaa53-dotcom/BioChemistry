# Jupyter Notebooks and Basic Navigation

Jupyter notebooks need two essential things to run.

1. A server in which the code is actually executed and
2. A browser window in which the notebook is shown.

## Ways of running the Jupyter Notebook Server

### Using Colab

We recommend to run the notebooks using the online service [CoLab](https://colab.research.google.com). Login requires a Google account, such as your LU student account.
When logged in, goto _Github_ and search for `luchem` and open the first notebook. Everything you do on Colab is **deleted** once you leave the session. To save your work, either download your notebook, or save it to Google Drive.

### Using a local Python Installation

If you have a local Anaconda installation on your laptop you should be good to go. To download Anaconda, see [Anaconda](https://www.anaconda.com/distribution/#download-section) or [miniconda](https://docs.conda.io/en/latest/miniconda.html) for a leaner installation.
The latter may require you to install additional conda pacakages.

### Using LU Desktop Computers
On the LU computers, we recommend to use Colab as described above in a browser. The computers also have an Anaconda installation and to use it, you need to start the Jupyter server in your folder. For this you need to add the python directory to your "path". The path are the places where windows looks to execute something. You have to do that only once. Either go to your user environment variable and add `c:Anaconda3\Scripts\` to it. (type this in the start menu)

Previous output can be retrieved by:
